```python
df = pd.read_csv('clustercad_db_all_full_modules_with_substrate.csv')
```

```python
df_annotations = pd.read_csv('clustercad_filtered.csv')
```

```python
import pandas as pd

merged_rows = []

for _, annotation_row in df_annotations.iterrows():
    kr_seq = annotation_row['kr_sequence']
    matching_row = df[df['protein_sequence'].str.contains(kr_seq, regex=False)]
    if not matching_row.empty:
        grab_one = matching_row.iloc[0]
        merged_row = {**annotation_row.to_dict(), **grab_one.to_dict()}
        merged_rows.append(merged_row)

merged_df = pd.DataFrame(merged_rows)

print(f"Merged DataFrame has {len(merged_df)} rows (should be <= {len(df_annotations)})")
merged_df.head()
```

```python
merged_df
```

```python
merged_df = merged_df.dropna()
```

```python
merged_df = merged_df.reset_index()
```

```python
unique_substrates = merged_df['substrate_annotation'].unique()
```

```python
unique_substrates
```

```python
# copy
filtered_df = merged_df.copy()

# case insensitive filtering
filtered_df['substrate_annotations_lower'] = filtered_df['substrate_annotation'].str.lower()

# "mmal" or "mal" filtering
mask_mmal = filtered_df['substrate_annotations_lower'].str.contains('mmal', na=False)
mask_mal = filtered_df['substrate_annotations_lower'].str.contains('mal', na=False) & ~mask_mmal

# filter
filtered_df = filtered_df[mask_mmal | mask_mal]

# substrate_type column
filtered_df['substrate_type'] = 'unknown'
filtered_df.loc[mask_mmal, 'substrate_type'] = 'mmal'
filtered_df.loc[mask_mal, 'substrate_type'] = 'mal'

filtered_df = filtered_df.drop(columns=['substrate_annotations_lower'])

print(f"Original DataFrame had {len(merged_df)} rows")
print(f"Filtered DataFrame has {len(filtered_df)} rows")
print(f"Found {mask_mmal.sum()} rows with 'mmal'")
print(f"Found {mask_mal.sum()} rows with 'mal' (excluding those with 'mmal')")

filtered_df
```

```python
def rewrite_annotations(kr_type, substrate_type):
    if substrate_type == 'mal':
        kr_type = kr_type[:-1]
    return (kr_type, substrate_type)
    
```

```python
array_kr = np.empty(0)
array_st = np.empty(0)

for row in filtered_df.itertuples():
    print(f"kr_type: {row.kr_type_annotation}, substrate_type: {row.substrate_type}")
    kr_type, kr_substrate = rewrite_annotations(row.kr_type_annotation, row.substrate_type)
    array_kr = np.append(array_kr, kr_type)
    array_st = np.append(array_st, kr_substrate)
```

```python
filtered_df['new_kr'] = array_kr
```

```python
filtered_df['new_st'] = array_st
```
